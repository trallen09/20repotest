```python
# import dependancies
import pandas as pd
```

```python
# read in csv
csv = pd.read_csv('Resources/Sale_Prices_City.csv')
sale_price_df = pd.DataFrame(csv)
sale_price_df.head()
```

```python
#find dtypes
sale_price_df.dtypes
```

```python
# filter data by state
tn_sale_df= sale_price_df.loc[sale_price_df['StateName'].isin(['Tennessee'])]
tn_sale_df.head()
```

```python
#drop unneeded columns and set index to Region Name
tn_sale_df = tn_sale_df.drop(columns=['Unnamed: 0', 'RegionID', 'StateName', 'SizeRank']).set_index('RegionName')
tn_sale_df
```

```python
#transpose dataframe to filter by dates
tn_transpose_df = tn_sale_df.T
tn_transpose_df
```

```python
# find number of null values by region
tn_transpose_df.isnull().sum()
```

```python
sales_df = (tn_transpose_df.loc['2015-01':'2020-03'])
sales_df
```

```python
#recount null values
sales_df.isnull().sum()
```

```python
#transpose dataframe back 
sales_df = sales_df.T
sales_df
```
